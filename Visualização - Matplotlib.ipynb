# Histogramas

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()
mediana = df['comprimento_pétala'].median()
eixo.set_title('Histograma', fontsize=14)
eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_xlabel('Comprimento da pétala', fontsize=12);
sns.set(style='darkgrid')
eixo.axvline(mu, color='k', linestyle='--')
eixo.axvline(sigma, color='g', linestyle='--')
eixo.axvline(sigma+mu, color='g', linestyle='--')
eixo.axvline(df['comprimento_pétala'].median(), color='r', linestyle='--')


eixo.annotate(f'média ({mu:.2f})', xy=(mu+0.1,30))
eixo.annotate(f'desvio {sigma:.2f}', xy=(sigma+0.1,30), c='g')
eixo.annotate(f'desvio {sigma+mu:.2f}', xy=(sigma+mu+0.1,30),c='g')
eixo.annotate(f'mediana {mediana:.2f}', xy=(mediana+0.1,28), color='r');
```

```python
# Treinando

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.set_title('Histograma da sépala', fontsize=14)
eixo.set_xlabel('Comprimento da sépala', fontsize=12)
eixo.set_ylabel('Quantidade de comprimentos', fontsize=12)
eixo.hist(x=df['comprimento_sépala'], bins=20)

media,mediana,desvio = df['comprimento_sépala'].mean(), df['comprimento_pétala'].median(), df['comprimento_pétala'].std()

eixo.axvline(media, color='r', linestyle='--');
eixo.axvline(mediana, color='g', linestyle='--')
eixo.axvline(media+desvio, color='cyan', linestyle='--')

eixo.annotate('média', xy=(media, 14))
eixo.annotate('mediana', xy=(mediana, 14))
eixo.annotate('desvio', xy=(media+desvio, 12))
```
